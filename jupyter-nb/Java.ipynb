## 1. Start Spark Context

```python
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .remote("sc://spark-connect:15002") \
    .getOrCreate()

# Enables Spark's native HTML rendering for Jupyter
spark.conf.set("spark.sql.repl.eagerEval.enabled", True)
spark.conf.set("spark.sql.repl.eagerEval.truncate", 255)

spark
```

## 2. Show Catalog Information

```python

spark.sql(""" show catalogs """)
```

```python

spark.sql(""" show schemas """)
```

```python

spark.sql(""" show tables in default """)
```

## 3. Load Data

```python

df = spark.read.format("parquet").load("file:///examples/java/java_vertices/")
df.createOrReplaceTempView("tmp_java_vertices")

spark.sql(""" drop table if exists default.java_vertices """)
spark.sql(""" CREATE TABLE IF NOT EXISTS default.java_vertices USING DELTA AS SELECT * FROM tmp_java_vertices """)
```

```python

spark.sql(""" select * from default.java_vertices """)
```

```python

df = spark.read.format("parquet").load("/examples/java/java_edges/")
df.createOrReplaceTempView("tmp_java_edges")

spark.sql(""" drop table if exists default.java_edges """)
spark.sql(""" CREATE TABLE IF NOT EXISTS default.java_edges USING DELTA AS SELECT * FROM tmp_java_edges """)
```

```python

spark.sql(""" select * from default.java_edges """)
```
